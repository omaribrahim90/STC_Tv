# top watched movies/series episodes based on :  Total Views / Total Users Views / Total watch time


Here we add the season/episode to the program name 

```python
df_classed=df.copy()
df_classed.loc[df_classed['program_class'] == 'SERIES/EPISODES', 'program_name'] = df_classed['program_name']+'_SE'+df_classed['season'].astype(str)+'_EP'+df_classed['episode'].astype(str)
```

```python
df_classed = df_classed.groupby(['program_name' , 'program_class']).agg({
    'user_id_maped': ['nunique' , 'count' ] ,
    'duration_seconds': 'sum' 
}).reset_index()

```

```python
df_classed.columns = ['program_name', 'program_class', 'No of Users watched', 'total watches', 'Total Watch time in sec']

df_classed['Total Watch time in hr'] = df_classed['Total Watch time in sec']/3600
```

```python
df_classed 
```

```python
df_classed = df_classed.drop(columns= 'Total Watch time in sec')
```

```python
df_classed.sort_values(by='Total Watch time in hr' , inplace=True , ascending=False)
```

```python
df_classed.head(20)
```

now we Visualize the result

```python
import matplotlib.pyplot as plt 
import seaborn as sns


```

```python
top10 = df_classed.nlargest(10, 'Total Watch time in hr')

plt.figure(figsize=(10, 8))
plt.pie(
    top10['Total Watch time in hr'],
    labels=top10['program_name'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Top 10 Programs by Total Watch Time (Hours)')
plt.axis('equal')  
plt.show()
```

# now we are going to see MOVIES vs SERIES based on : Total Views / Total Users Views / Total watch time

```python
comparison = df.copy()
comparison = comparison.groupby('program_class').agg({
    'user_id_maped': ['nunique' , 'count' ] ,
    'duration_seconds': 'sum' 
}).reset_index() 


```

```python
comparison.columns = ['program_class' , 'No of Users watched', 'total watches', 'Total Watch time in sec']
comparison['Total Watch time in hr'] = comparison['Total Watch time in sec']/3600
comparison = comparison.drop(columns= 'Total Watch time in sec')
comparison.sort_values(by='Total Watch time in hr' , inplace=True , ascending=False)
```

```python
comparison
```

```python
comparison.set_index('program_class')['Total Watch time in hr'].plot.pie(
    autopct='%1.1f%%',  
    startangle=90,       
    figsize=(6, 6),        
    ylabel='',            
    title='Watch Time distribution by Program Class'
)

plt.tight_layout()
plt.show()
```

```python
comparison.set_index('program_class')['No of Users watched'].plot.pie(
    autopct='%1.1f%%',  
    startangle=90,       
    figsize=(6, 6),        
    ylabel='',            
    title='Number of users distribution by Program Class'
)

plt.tight_layout()
plt.show()
```
